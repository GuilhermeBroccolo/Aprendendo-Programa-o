# Código para calcular o preço para um Peça em 3D #

```python
def calcular_custo_filamento():
    while True:
        try:
            custo_filamento = float(input("Digite a quantidade(g) de filamento gasto: "))
            if custo_filamento > 0:
                print(f"\n\033[32mA Quantidade de {custo_filamento:.2f} foi registrado! \033[0m")
                return custo_filamento
            else:
                print("\n\033[31mA Quantidade não pode ser Zero ou Negativo! \n\033[0m")
        except ValueError:
            print("\n\033[31mDigite apenas a quantidade Certa!\n\033[0m")


gramas = calcular_custo_filamento()

preco_kg = 120
preco_por_grama = preco_kg / 1000
custo_final = gramas * preco_por_grama

print(f"Peso utilizado: {gramas:.2f}")
print(f"Custo total: R$ {custo_final:.2f}")
print(calcular_custo_filamento)
```
